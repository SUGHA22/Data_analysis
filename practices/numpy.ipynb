```python
import numpy as np
arr=np.array([1,2,3])
print("number of dimensional:",arr.ndim)
print("size of array :",arr.size)
print("data type:",arr.dtype)
print("shape of array:",arr.shape)
print("item size",arr.itemsize)
arr1=np.array([[1,2,3],[2,3,4]])
print("number of dimensional:",arr1.ndim)
print("size of array :",arr1.size)
print("data type:",arr1.dtype)
print("shape of array:",arr1.shape)
print("item size",arr1.itemsize)
arr2=np.array([[[1,2,3],
               [1,2,3],
               [4,5,6]]])

print("number of dimensional:",arr2.ndim)
print("size of array :",arr2.size)
print("data type:",arr2.dtype)
print("shape of array:",arr2.shape)
print("item size",arr2.itemsize)
```
