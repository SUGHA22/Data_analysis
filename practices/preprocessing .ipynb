```python
col_removed=df.drop(columns=['song_popularity'])
col_removed
```

```python
arr=pd.concat([col_removed,df['song_popularity']],axis=1)
arr
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
for i in arr.columns:
 arr[i]=sc.fit_transform(arr[i].values.reshape(-1,1))
arr
```

```python
import matplotlib.pyplot as plt
for i in arr.columns:
 c=plt.boxplot(arr[i])
 plt.title(i)
 plt.show()
```

```python
def drop_outliers(df,columns):
  df_filtered=df.copy()
  for column in columns:
    Q1=df[column].quantile(0.25)
    Q3=df[column].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5* IQR
    upper_bound = Q3 + 1.5 * IQR


    df_filtered = df_filtered[(df_filtered[column] >= lower_bound) & (df_filtered[column] <= upper_bound)]

    return df_filtered




df_filtered = drop_outliers(df,arr)

print("Original DataFrame:")
print(df)
print("\nDataFrame after removing outliers:")
print(df_filtered)
```

```python
import matplotlib.pyplot as plt
for i in df_filtered.columns:
 c=plt.boxplot(df_filtered[i])
 plt.title(i)
 plt.show()
```
